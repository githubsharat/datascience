1. **Key Tasks in Machine Learning**:
    - Data Collection
    - Data Preprocessing
    - Feature Engineering
    - Model Selection
    - Model Training
    - Evaluation
    - Deployment
    **Data Pre-processing**: The process of cleaning and transforming raw data into a usable format for machine learning.

2. **Quantitative and Qualitative Data**:
    - **Quantitative**: Deals with numbers and can be measured. E.g., age, height.
    - **Qualitative**: Descriptive data that can be categorized but not measured. E.g., color, gender.
    **Distinction**: Quantitative is numeric and measurable, while qualitative is descriptive and categorical.

3. **Sample Data Collection**:
    | Name | Age (Quantitative) | Gender (Qualitative) | Customer Rating (Ordinal) |
    |------|-------------------|----------------------|--------------------------|
    | John | 25                | Male                 | Good                     |
    | Anna | 30                | Female               | Excellent                |

4. **Causes of ML Data Issues**:
    - Missing values
    - Outliers
    - Inconsistencies
    - Duplicate records
    **Ramifications**: Poor model performance, misleading results, overfitting.

5. **Exploration of Categorical Data**:
    - **Frequency Distribution**: Counting occurrences of each category. E.g., the number of males and females in a dataset.
    - **Bar Charts**: Visual representation of frequency distribution.
    - **Cross-tabs**: Contingency tables displaying the relationship between two categorical variables.

6. **Effect of Missing Values**: It can lead to reduced model accuracy, biased results, and overfitting.
   **Handling**: Imputation, deletion, or using algorithms robust to missing data.

7. **Handling Missing Data**:
    - **Mean/Median Imputation**: Filling missing values with the mean or median.
    - **Mode Imputation**: For categorical data.
    - **Prediction Models**: Using algorithms like k-NN.
    - **Deletion**: Removing rows with missing values.

8. **Data Pre-processing Techniques**:
    - Normalization
    - Standardization
    - One-hot encoding
    - Binning
    **Dimensionality Reduction**: Reducing the number of input variables.
    **Feature Selection**: Choosing the most important variables.

9. **i. IQR**: Measures the statistical spread between the 25th and 75th percentile. 
   **Criteria**: Difference between the first (Q1) and third quartile (Q3).
   **ii. Box Plot Components**:
    - Median: Middle value.
    - Upper Whisker: Q3 + 1.5 * IQR.
    - Lower Whisker: Q1 - 1.5 * IQR.
    - Box: Represents Q1 to Q3.
    **Lower whisker > Upper whisker**: When there's more spread below Q1 than above Q3.
    **Identifying Outliers**: Data points outside the whiskers.

10. **Notes**:
    **1. Data Collected at Regular Intervals**: Time-series data, like stock prices.
    **2. Gap Between Quartiles**: Reflects spread in the middle 50% of the data.
    **3. Cross-tab**: Contingency tables showing relation between two categorical variables.

11. **Comparison**:
    **1. Nominal vs. Ordinal Data**: Nominal is categorical without order (e.g., color). Ordinal has a clear ordering (e.g., rating: poor, good, excellent).
    **2. Histogram vs. Box Plot**: Histogram shows data distribution, while box plot shows statistical spread and outliers.
    **3. Mean vs. Median**: Mean is the average; median is the middle value. Median is less sensitive to outliers.

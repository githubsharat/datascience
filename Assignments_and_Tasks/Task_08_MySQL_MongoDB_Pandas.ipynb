#### Import packages and connect to MySQL

```python
import mysql.connector as conn
import pymongo
import json
import csv
import pandas as pd

mydb = conn.connect(host = "localhost" , user ="root" , passwd = "root" )
cursor = mydb.cursor()
```

#### Create a database and use it

```python
s1 = "create database if not exists db1"
cursor.execute(s1)
s1 = "use db1"
cursor.execute(s1)
```

#### 1. Create tables for attribute dataset and dress dataset

```python
s1 = "create table if not exists db1.attribute_table(Dress_ID varchar(10), Style varchar(10), Price varchar(10), Rating varchar(10), Size varchar(10), Season varchar(10), NeckLine varchar(25), SleeveLength varchar(25), waiseline varchar(25), Material varchar(25), FabricType varchar(25), Decoration varchar(25),Pattern_Type varchar(25), Recommendation varchar(10))"
q1 = cursor.execute(s1)

s1 = "create table if not exists db1.sales_table(Dress_ID varchar(25),`29/8/2013` varchar(25),`31/8/2013` varchar(25),`2013-02-09` varchar(25),`2013-04-09` varchar(25),`2013-06-09` varchar(25),`2013-08-09` varchar(25),`2013-10-09` varchar(25),`2013-12-09` varchar(25),`14/9/2013` varchar(25),`16/9/2013` varchar(25),`18/9/2013` varchar(25),`20/9/2013` varchar(25),`22/9/2013` varchar(25),`24/9/2013` varchar(25),`26/9/2013` varchar(25),`28/9/2013` varchar(25),`30/9/2013` varchar(25),`2013-02-10` varchar(25),`2013-04-10` varchar(25),`2013-06-10` varchar(25),`2010-08-10` varchar(25),`2013-10-10` varchar(25),`2013-12-10` varchar(25))"
q1 = cursor.execute(s1)
```

#### 2. Bulk load the data (from datasets) into the two tables

```python
csv_data = csv.reader(open('Attribute_DataSet.csv'))
header = next(csv_data)
for row in csv_data:
    cursor.execute("insert into attribute_table(Dress_ID, Style, Price, Rating, Size, Season, NeckLine, SleeveLength, waiseline, Material, FabricType, Decoration, Pattern_Type, Recommendation) values (%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)", row)
mydb.commit()

csv_data = csv.reader(open('DressSales.csv'))
header = next(csv_data)
for row in csv_data:
    cursor.execute("insert into sales_table(Dress_ID ,`29/8/2013` ,`31/8/2013` ,`2013-02-09` ,`2013-04-09` ,`2013-06-09` ,`2013-08-09` ,`2013-10-09` ,`2013-12-09` ,`14/9/2013` ,`16/9/2013` ,`18/9/2013` ,`20/9/2013` ,`22/9/2013` ,`24/9/2013` ,`26/9/2013` ,`28/9/2013` ,`30/9/2013` ,`2013-02-10` ,`2013-04-10` ,`2013-06-10` ,`2010-08-10` ,`2013-10-10` ,`2013-12-10`) values (%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s, %s)", row)
mydb.commit()
```

#### 3. Create two Dataframes using data in the two tables

```python
cursor.execute('use db1')

cursor.execute('SELECT * FROM attribute_table')
table_rows = cursor.fetchall()

df_attribute = pd.DataFrame(table_rows)
df_attribute.columns = ['Dress_ID','Style','Price','Rating','Size','Season','NeckLine','SleeveLength','Waiseline','Material','FabricType','Decoration','Pattern Type','Recommendation']


cursor.execute('SELECT * FROM sales_table')
table_rows = cursor.fetchall()

df_sales = pd.DataFrame(table_rows)
df_sales.columns = ['Dress_ID','29/8/2013','31/8/2013','2013-02-09','2013-04-09','2013-06-09','2013-08-09','2013-10-09','2013-12-09','14/9/2013','16/9/2013','18/9/2013','20/9/2013','22/9/2013','24/9/2013','26/9/2013','28/9/2013','30/9/2013','2013-02-10','2013-04-10','2013-06-10','2010-08-10','2013-10-10','2013-12-10']
```

#### 7. Write a sql query to find out how many unique dress that we have based on dress id

```python
mydb = conn.connect(host = "localhost" , user ="root" , passwd = "root" )
cursor = mydb.cursor()

cursor.execute("use db1")
s1 = 'select count(distinct(Dress_ID)) from attribute_table'
cursor.execute(s1)
print(cursor.fetchall()[0][0])
```

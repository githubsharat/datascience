We can solve the equations either algebraically or numerically.

Let's solve them alebraically first

Let's take the following two example equations

`2x + 5y = 12`   **(Equation 1)**

`x + y = 3`      **(Equation 2)**

`2x + 5y = 12`     
`2x + 2y = 6`   (Multiply equation 2 with 2)


`3y = 6`        (subtracting the two) 

`y = 2`     
`x = 1`

The solution for the two equations is `(1,2)`

#### Let's solve the equations numerically using the Power of Computers!

In order to solve the above two equation numerically, we will solve the equations through iterations     

so, we need to create a new equation with the following structure: `y = f(y)` or `x = f(x)` that will help us set up the iterations    

In order to obtain that structure, we need to rearrange terms.    

Let's re-arrange Equation 2 as: `x = 3 - y`. Let's call this as **Equation (3)** 

Let's re-arrange Equation 1 as: `5y = 12 - 2x`. Let's call this as **Equation (4)**

Substituting the value of `x` from Equation (3) in Equation (4), we get...

`5y = 12 - 2 (3 - y)`

`5y = 12 - 6 + 2y`

`5y = 6 + 2 y`

`y = (6 + 2 y) / 5`

**Therefore, we have acheived `y = f(y)` structure - Highly desired to solve numerically**

Note down this equation: `y = (6 + 2 y) / 5`

Let's solve the equation numerically through iterations

For the first iteration, we need the guess the value of y randomly.

We can use `NumPy` library to generate a random number for the value of y from, for example,a -999999 to +999999: 

`numpy.random.randint(-999999, 999999)`

This will generate either a `positive number` , `negative number`, or `zero` in the range of (-999999, 999999)
